```python
1;LC;85387;C;6;Rosa;457;0;0
2;LC;87507;B;6;Rosa;457;0;0
3;LC;13652;E;7;Rosa;457;0;0
4;LC;95804;D;8;Rosa;457;0;0
5;LC;87035;B;5;Rosa;457;0;0
1;LC;84836;B;5;Rosa;457;1;0
2;LC;59531;B;7;Rosa;457;1;0
3;LC;55925;D;6;Rosa;457;1;0
4;LC;32893;C;5;Rosa;457;1;0
5;LC;88480;B;5;Rosa;457;1;0
```

```python
comp = 'LC'

item_ing = ['Item 85387', 'Item 87507', 'Item 13652', 'Item 95804', 'Item 87035']
item_esp = ['Item 84836', 'Item 59531','Item 55925', 'Item 32893', 'Item 88480']

df_reg = get_df_reg(comp, ano)
df_reg_ing = {}
df_reg_esp = {}
df_reg_pt = {}

for reg in df_reg.keys():
    df_reg_ing[reg] = df_reg[reg][df_reg[reg]['TP_LINGUA']==0][['NU_INSCRICAO','NU_NOTA_LC']].copy()
    df_reg_esp[reg] = df_reg[reg][df_reg[reg]['TP_LINGUA']==1][['NU_INSCRICAO','NU_NOTA_LC']].copy()
    df_reg_pt[reg] = df_reg[reg][['NU_INSCRICAO','NU_NOTA_LC']].copy()
    
    for col in df_reg['N'].columns:
        if col in item_ing:
            df_reg_ing[reg][col] = df_reg[reg][df_reg[reg]['TP_LINGUA']==0][col]
        elif col in item_esp:
            df_reg_esp[reg][col] = df_reg[reg][df_reg[reg]['TP_LINGUA']==1][col]
        elif col!='NU_INSCRICAO' and col!='NU_NOTA_LC' and col!='TP_LINGUA':
            df_reg_pt[reg][col] = df_reg[reg][col]
    
```

```python
print('Média e Std para Inglês:')
print_mean_std(df_reg_ing, comp)
```

```python
print('Média e Std para Espanhol:')
print_mean_std(df_reg_esp, comp)
```

```python
print('Média e Std total:')
print_mean_std(df_reg_pt, comp)
```

```python
plot_hist_scores(df_reg_pt, comp)
```
